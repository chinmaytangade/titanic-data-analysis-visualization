```python
# ===============================
#  Titanic Data Analysis Script
# ===============================

# --- Import Libraries ---
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# --- Display Settings ---
pd.set_option('display.max_columns', None)
plt.style.use('seaborn-v0_8')

# --- Load Dataset ---
#  Make sure the file path below is correct for your system
df = pd.read_csv(r"C:\Users\chinm\OneDrive\Desktop\chinmayT\internship1\train.csv")


# --- Basic Info ---
print("Shape of dataset:", df.shape)
print("\nData Info:\n")
print(df.info())
print("\nSummary Statistics:\n")
print(df.describe())
print("\nMissing Values:\n")
print(df.isnull().sum())

# --- Univariate Analysis ---
print("\n=== Univariate Analysis ===")

# Histograms for numeric columns
df.hist(figsize=(10, 8), color='skyblue', edgecolor='black')
plt.suptitle("Numeric Feature Distributions", fontsize=14)
plt.tight_layout()
plt.show()

# Boxplots
for col in ['Age', 'Fare']:
    plt.figure()
    sns.boxplot(x=df[col], color='lightgreen')
    plt.title(f'Boxplot of {col}')
    plt.show()

# Categorical counts
plt.figure()
sns.countplot(x='Sex', data=df)
plt.title('Gender Distribution')
plt.show()

plt.figure()
sns.countplot(x='Survived', data=df)
plt.title('Survival Count (0 = No, 1 = Yes)')
plt.show()

# --- Bivariate Analysis ---
print("\n=== Bivariate Analysis ===")

plt.figure()
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title('Survival by Gender')
plt.show()

plt.figure()
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title('Survival by Passenger Class')
plt.show()

plt.figure()
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=df)
plt.title('Age vs Fare (Colored by Survival)')
plt.show()

# --- Correlation Heatmap ---
print("\n=== Correlation Heatmap ===")
corr = df.corr(numeric_only=True)
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()

# --- Pairplot ---
print("\n=== Pairplot ===")
sns.pairplot(df[['Survived', 'Pclass', 'Age', 'Fare']], hue='Survived', diag_kind='kde')
plt.show()

# --- Summary of Findings ---
print("\n=== Summary of Key Insights ===")
print("""
1 Most passengers were from 3rd class.
2 Females had higher survival rates than males.
3 Passengers in 1st class had greater chances of survival.
4 Higher fares corresponded with higher survival probability.
5 Age had mild effect: younger passengers survived slightly more.
""")
```
